## Model Validation

Banyak orang membuat kesalahan besar saat mengukur akurasi prediksi. Membuat prediksi dengan data pelatihan mereka dan membandingkan prediksi tersebut dengan nilai target dalam data pelatihan. Melihat masalah dengan pendekatan ini dan cara mengatasinya

```python
import pandas as pd
```

```python
# Load data
data = pd.read_csv ("/content/melb_data.csv")
```

```python
data
```

```python
# Filter rows with missing price values
filtered_data = data.dropna(axis=0)
```

Memfilter DataFrame dan menghapus baris dengan nilai yang hilang di kolom 'harga'

```python
 # Choose target and features
y = filtered_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea',
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = filtered_data[melbourne_features]
```

y = sebagai kolom 'Harga' dari filtered_dataDataFrame
X = berisi fitur tertentu dari filtered_dataDataFrame


```python
from sklearn.tree import DecisionTreeRegressor
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(X, y)
```

Impor DecisionTreeRegressor mengimpor DecisionTreeRegressorkelas dari sklearn.tree

fit metode objek melbourne_model, meneruskan fitur ( X) dan variabel target ( y) sebagai argumen.

model DecisionTree pada data yang disediakan, mempelajari hubungan antara fitur dan variabel target.

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

Impor mean_absolute_error mengimpor mean_absolute_errorfungsi dari sklearn.metrics digunakan untuk menghitung rata-rata kesalahan absolut antara dua kumpulan nilai.

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

Impor train_test_split mengimpor train_test_splitfungsi dari sklearn.model_selection digunakan untuk membagi kumpulan data menjadi kumpulan pelatihan dan validasi.

train_test_split meneruskan fitur ( X), variabel target ( y), dan random_state argumen

train_X: Fitur pelatihan
val_X: Fitur validasi
val_X: Fitur validasi
val_y: Variabel target validasi
