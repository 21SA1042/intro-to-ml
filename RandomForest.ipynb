## Random Forest

- Menggunakan banyak decision tree.
- Membuat prediksi dengan menghitung rata-rata prediksi dari setiap pohon.
- Umumnya memiliki akurasi prediksi yang jauh lebih baik daripada pohon keputusan tunggal.
- Bekerja dengan baik menggunakan parameter default.
- Dengan pemodelan lebih lanjut, model dengan kinerja lebih baik dapat ditemukan, namun banyak dari model tersebut sensitif terhadap pengaturan parameter yang tepat.


```python
import pandas as pd
```

```python
# Load data
data = pd.read_csv ("/content/melb_data.csv")
```

```python
# Filter rows with missing values
data = data.dropna(axis=0)

# Choose target and features
y = data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea',
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = data[melbourne_features]
```

dropna digunakan dengan axis=0   menghapus baris (sumbu 0) dengan nilai yang hilang
- y: Variabel ini menyimpan kolom 'Harga', yang mewakili variabel target yang ingin kita prediksi atau modelkan
- X: DataFrame ini berisi kolom sisa yang ditentukan dalam melbourne_featuresdaftar. Kolom ini mewakili fitur yang akan kita gunakan untuk memprediksi variabel target.

DataFrame yang dihasilkan filtered_datahanya berisi baris yang semua kolomnya memiliki nilai valid


```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

Mengimpor train_test_split fungsi dari sklearn.model_selection digunakan untuk membagi kumpulan data menjadi kumpulan pelatihan dan validasi.

train_test_split: Fungsi ini membutuhkan beberapa argumen
- X: Matriks fitur yang berisi titik data (baris mewakili sampel, kolom mewakili fitur)
- y: Vektor variabel target (nilai yang ingin kita prediksi)
- random_state = 0 :  menetapkan benih untuk penghasil angka acak yang digunakan dalam proses pemisahan

train_X: Fitur set pelatihan
val_X: Fitur set validasi
train_y: Variabel target set pelatihan
val_y: Variabel target set validasi




```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

RandomForestRegressor sklearn.ensemble digunakan untuk membuat dan melatih model regresi Random Forest

mean_absolute_error sklearn.metrics digunakan untuk menghitung mean absolute error (MAE) antara dua kumpulan nilai

fit forest_modeluntuk melatih model Random Forest pada data pelatihan. Mempelajari hubungan antara fitur ( train_X) dan variabel target ( train_y)

(val_X) menggunakan model Random Forest yang dilatih. Nilai prediksi disimpan dalam variabel melb_preds

( val_y) dan harga prediksi ( melb_preds) menggunakan mean_absolute_error untuk menghitung kesalahan absolut rata-rata yang dihitung kemudian dicetak.

- MAE adalah metrik yang digunakan untuk mengukur perbedaan absolut rata-rata antara nilai prediksi dan nilai sebenarnya
- Dengan mengevaluasi model pada data validasi mendapatkan penilaian yang lebih realistis mengenai kemampuan generalisasi dan potensi kinerjanya pada data baru yang belum terlihat
- Membantu menghindari overfitting dan memilih model terbaik untuk masalah spesifik


Model Random Forest menunjukkan peningkatan kinerja yang signifikan dibandingkan dengan model Decision Tree,  tanpa penyetelan parameter lebih lanjut
